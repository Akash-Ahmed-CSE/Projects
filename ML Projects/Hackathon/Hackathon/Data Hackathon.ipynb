```python
Features = ['Age', 'FCV19S_1', 'FCV19S_2', 'FCV19S_3', 'FCV19S_4', 'FCV19S_5', 'FCV19S_6', 'FCV19S_7', 'PHQ9_1','PHQ9_2', 'PHQ9_3', 'PHQ9_4', 'PHQ9_5', 'PHQ9_6', 'PHQ9_7', 'PHQ9_8', 'PHQ9_9', 'PHQ9_CATEGORY', 'ISI_1', 'ISI_2', 'ISI_3', 'ISI_4', 'ISI_5', 'ISI_6', 'ISI_7', 'ISI_CATEGORY']
Y_TrainData = df_train['Suicide']
X_TrainData = df_train[Features]
X_TestData = df_test[Features]
X_TrainData
```

```python
X_TestData
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
X_TrainData = preprocessing.StandardScaler().fit(X_TrainData).transform(X_TrainData)
X_TestData = preprocessing.StandardScaler().fit(X_TestData).transform(X_TestData)
```

```python
xgb = XGBClassifier()
xgb.fit(X_TrainData, Y_TrainData)
Y_pred_xgb = xgb.predict(X_TestData)
```

```python
res = pd.DataFrame(Y_pred_xgb)
```
